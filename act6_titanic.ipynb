<h1>BASIC OF MACHINE LEARNING</h1>

```python
import pandas as pd 
```

```python
df = pd.read_csv('Titanic.csv') 
```

```python
df
```

```python
# Drop irrelevant columns
df = df[["Survived", "Pclass", "Sex", "Age","Embarked"]]

# Drop missing values
df = df.dropna()

# Encode categorical variable
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({"C":0, "Q":1, "S":2})
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[["Pclass", "Sex", "Age", "Embarked"]]
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

<H1>ACCURACY TEST</H1>

```python

from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

<h1> Accuracy Result</h1>

```python
# Keep only useful columns

df2 = pd.read_csv("Titanic.csv")
df2
```

```python
# Keep only useful columns
df2 = df2[["Survived", "Pclass", "Sex", "Age", "Fare", "Embarked"]]

# Drop missing values
df2 = df2.dropna()
```

```python
df2
```

```python
# Encode 'Sex' and 'Embarked'
df2["Sex"] = df2["Sex"].map({"male": 0, "female": 1})
df2["Embarked"] = df2["Embarked"].map({"S": 0, "C": 1, "Q": 2})
```

```python
df2
```

```python
X = df2[["Pclass", "Sex", "Age", "Fare", "Embarked"]]
y = df2["Survived"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

<h1>Manual Testing</h1>

```python

passenger = pd.DataFrame({'Pclass': [1],'Sex': [0],'Age': [22],'Fare': [7.25],'Embarked': [0]})

prediction = model.predict(passenger)[0]

if prediction == 1:
    print("Passenger likely survive")
else:
    print("Passenger unlikely to survive")
```
